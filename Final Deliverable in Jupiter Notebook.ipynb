# Concatenating season with name to create a primary/foreign key

```python
filename2['id2'] = filename2['Season'].astype(str)+ '_'+ filename2['Name']
filename2
```

```python
filename2['id2'] = filename2['id2'].str.upper()
filename2
```

# Seperate the id from last name to further make a primary key

```python
filename2[['id2','last name']] = filename2['id2'].str.split(n = 1,expand=True)
```

# Replace the the last name that says none to X so we could make a primary key

```python

filename2["last name"].replace({None: "X"}, inplace=True)
```

# Took just the first letter of the last name so we could make a primary key

```python
filename2['first letter'] = filename2['last name'].astype(str).str[0]
filename2
```

```python

#filename2['id2'] = filename2['Season'].astype(str)+ '_'+ filename2['Name']
filename2['contestant id'] = filename2['id2'].astype(str)+ '_'+filename2['first letter']
filename2
```
